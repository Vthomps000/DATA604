A random seed is a number or vector that initializes a pseudorandom number generator. A pseudorandom number generator's number sequence is determined completely by the seed. Because of the nature of number generating algorithms, so long as the original seed is ignored, the rest of the values generated by the algorithm will follow pseudorandom probability distribution 
Simulation studies are computer experiments that involve creating data by pseudo‐random sampling from known probability distributions. They are an invaluable tool in statistical research that evaluate new methods and compare alternative methods. A key strength of simulation is the ability to understand the behavior of statistical methods since some truth from parameters of interest is known from the process of generating the data. This allows us to consider properties of methods, like bias. Proper understanding of simulation studies enable statisticians to both run and critically appraise published simulation studies, and conduct simulation studies with greater care and report with transparency.
